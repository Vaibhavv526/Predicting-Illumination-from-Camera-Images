```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
import os

bright_ids = {
    os.path.splitext(f)[0]
    for f in os.listdir("data/train/bright")
}

dark_ids = {
    os.path.splitext(f)[0]
    for f in os.listdir("data/train/dark")
}

normal_ids = {
    os.path.splitext(f)[0]
    for f in os.listdir("data/train/normal")
}
```

```python
print("Bright folder labels:")
print(df[df["id"].isin(bright_ids)]["label"].value_counts())

print("\nDark folder labels:")
print(df[df["id"].isin(dark_ids)]["label"].value_counts())

print("\nNormal folder labels:")
print(df[df["id"].isin(normal_ids)]["label"].value_counts())
```
