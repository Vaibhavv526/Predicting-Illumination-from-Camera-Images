```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

resnet_df = pd.read_csv("resnet_features.csv")
yolo_df = pd.read_csv("yolo_features.csv")
handcrafted_df = pd.read_csv("handcrafted_features.csv")

print("ResNet Shape      :", resnet_df.shape)
print("YOLO Shape        :", yolo_df.shape)
print("Handcrafted Shape :", handcrafted_df.shape)
```

```python
print("ResNet Columns:")
print(resnet_df.columns.tolist())

print("\nYOLO Columns:")
print(yolo_df.columns.tolist())

print("\nHandcrafted Columns:")
print(handcrafted_df.columns.tolist())
```

problem is that yolo and handcrafted csv have id but resnet have images not id so we have to rename it as id

```python
# Standardize image filename column
yolo_df = yolo_df.rename(columns={"id": "image"})
handcrafted_df = handcrafted_df.rename(columns={"id": "image"})
```

```python
print("Image Alignment")
print("ResNet == YOLO       :", (resnet_df["image"] == yolo_df["image"]).all())
print("ResNet == Handcrafted:", (resnet_df["image"] == handcrafted_df["image"]).all())

print("\nLabel Alignment")
print("ResNet == YOLO       :", (resnet_df["label"] == yolo_df["label"]).all())
print("ResNet == Handcrafted:", (resnet_df["label"] == handcrafted_df["label"]).all())
```

```python
comparison = pd.DataFrame({
    "ResNet": resnet_df["image"],
    "YOLO": yolo_df["image"]
})

comparison.head(20)
```

```python
print("First 10 ResNet images:")
print(resnet_df["image"].head(10).tolist())

print("\nFirst 10 YOLO images:")
print(yolo_df["image"].head(10).tolist())
```

```python
resnet_df["image"] = resnet_df["image"].astype(str) + ".png"
```

```python
print((resnet_df["image"] == yolo_df["image"]).all())
```

```python
print("ResNet unique images:", resnet_df["image"].nunique())
print("YOLO unique images:", yolo_df["image"].nunique())

print("\nImages missing in YOLO:")
print(len(set(resnet_df["image"]) - set(yolo_df["image"])))

print("\nImages missing in ResNet:")
print(len(set(yolo_df["image"]) - set(resnet_df["image"])))
```

resnet images don't have .png but in yolo it have .png so we have to rename resnet to .png


```python
resnet_df["image"] = (
    resnet_df["image"]
    .str.replace(r"(\.png)+$", "", regex=True)
    + ".png"
)
yolo_df = yolo_df.rename(columns={"id": "image"})
handcrafted_df = handcrafted_df.rename(columns={"id": "image"})
```

```python
print("ResNet label dtype:", resnet_df["label"].dtype)
print("YOLO label dtype:", yolo_df["label"].dtype)
print("Handcrafted label dtype:", handcrafted_df["label"].dtype)

print("\nUnique labels:")
print("ResNet:", sorted(resnet_df["label"].unique()))
print("YOLO:", sorted(yolo_df["label"].unique()))
print("Handcrafted:", sorted(handcrafted_df["label"].unique()))
```

```python
label_mapping = {
    "dark": 0,
    "normal": 1,
    "bright": 2
}

if yolo_df["label"].dtype == object:
    yolo_df["label"] = yolo_df["label"].map(label_mapping)
```

```python
yolo_df["label"] = (
    yolo_df["label"]
    .astype(str)
    .str.strip()
    .str.lower()
)

label_mapping = {
    "dark": 0,
    "normal": 1,
    "bright": 2
}

yolo_df["label"] = yolo_df["label"].map(label_mapping)
```

### merging all 

```python
merged_df = pd.merge(
    resnet_df,
    yolo_df,
    on=["image", "label"],
    how="inner"
)
```

```python
# image + integer label
final_df = pd.merge(
    merged_df,
    handcrafted_df,
    on=["image", "label"],
    how="inner"
)
```
